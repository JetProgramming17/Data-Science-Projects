# Apollo Mokoena's Investment Strategy

Welcome to the investment simulation and forecasting model for Apollo Mokoena, a 20-year-old aspiring data scientist from Johannesburg, South Africa. This notebook walks through the process of analyzing different exchange-traded funds (ETFs), simulating a balanced portfolio, applying tax considerations, and forecasting future wealth.


## üìå Objectives

- Simulate the historical growth of a R100,000 ETF-based portfolio
- Apply South African capital gains tax to understand net investment value
- Forecast the next 10 years of portfolio growth using linear regression
- Evaluate assumptions and include environmental market analysis
- Recommend the most strategic mix of ETFs for Apollo based on risk, return, and global diversification


## üß† Assumptions

1. Apollo invests R100,000 once, with no monthly contributions.
2. The ETFs selected are broad-market and historically diversified.
3. Portfolio is rebalanced only through normalized weights ‚Äî no active trading.
4. South African Capital Gains Tax of 18% is applied only at liquidation (end of investment horizon).
5. Inflation and dividends are not included in this simulation (but can be modeled in Part B).
6. Data source: Yahoo Finance via yfinance API.


## üåç Environmental and Market Context

### South Africa (JSE ETFs)
- STX40 and STXDIV offer exposure to top South African companies and dividend-focused portfolios.
- STXPRO (property ETF) underperformed in recent years due to interest rate hikes and post-COVID real estate softness.
- STXWDM provides ZAR-based global equity exposure with foreign diversification.

### International Markets
- VOO and VT track S&P 500 and global stock performance respectively ‚Äî generally more stable and higher long-term returns.
- EEM targets emerging markets (Brazil, India, China) ‚Äî more volatile but potential for higher growth.

‚ö†Ô∏è Past performance is not indicative of future results. Global inflation, recession fears, and geopolitical instability may impact returns.


```python
weights = {
    "STX40": 0.20,
    "STXDIV": 0.10,
    "STXWDM": 0.20,
    "STXPRO": 0.10,
    "VOO": 0.20,
    "VT": 0.10,
    "EEM": 0.10
}
```

```python
tickers = list(weights.keys())
df_portfolio = df_closes[tickers].dropna()
normalized = df_portfolio / df_portfolio.iloc[0]

for ticker, weight in weights.items():
    normalized[ticker] = normalized[ticker] * weight

portfolio_value = normalized.sum(axis=1) * 100_000
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
portfolio_value.plot(label='Apollo Portfolio', color='green')
plt.title("Apollo's R100,000 Portfolio Growth (2013‚ÄìNow)")
plt.xlabel("Date")
plt.ylabel("Portfolio Value (ZAR)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
initial_value = 100_000
final_value = portfolio_value.iloc[-1]
capital_gain = final_value - initial_value
capital_gains_tax = capital_gain * 0.18
after_tax_value = final_value - capital_gains_tax

import pandas as pd
summary_df = pd.DataFrame({
    "Initial Investment (ZAR)": [initial_value],
    "Final Value (Pre-Tax)": [round(final_value, 2)],
    "Capital Gain": [round(capital_gain, 2)],
    "Capital Gains Tax (18%)": [round(capital_gains_tax, 2)],
    "Final Value (After-Tax)": [round(after_tax_value, 2)]
})
summary_df.T

```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

portfolio_df = pd.DataFrame({'Date': portfolio_value.index, 'Value': portfolio_value.values})
portfolio_df['Ordinal'] = portfolio_df['Date'].map(pd.Timestamp.toordinal)

X = portfolio_df[['Ordinal']]
y = portfolio_df['Value']
model = LinearRegression()
model.fit(X, y)

# Forecast 10 years ahead
future_dates = pd.date_range(start=portfolio_df['Date'].iloc[-1], periods=120, freq='MS')
future_df = pd.DataFrame({'Date': future_dates})
future_df['Ordinal'] = future_df['Date'].map(pd.Timestamp.toordinal)
future_df['Forecasted_Value'] = model.predict(future_df[['Ordinal']])
```

```python
plt.figure(figsize=(14, 7))
plt.plot(portfolio_df['Date'], portfolio_df['Value'], label='Historical (Pre-Tax)', color='blue')
plt.plot(future_df['Date'], future_df['Forecasted_Value'], label='Forecast (Next 10 Years)', linestyle='--', color='orange')
plt.title("Apollo's Portfolio: Historical and 10-Year Forecast")
plt.xlabel("Date")
plt.ylabel("Portfolio Value (ZAR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ‚úÖ Synthesis and Final Recommendation

After analyzing 7 ETFs over a 10-year period and forecasting 10 years forward:

- Apollo's portfolio grows steadily with a compound annual growth rate (CAGR) above 8% (pre-tax).
- International exposure (VOO, VT, STXWDM) outperforms local ETFs in both return and stability.
- The optimal portfolio mix combines both South African and international ETFs for diversification, risk balancing, and tax optimization via STXWDM (ZAR-based global exposure).

### üîí Recommendation:
> Apollo should maintain a medium-risk portfolio with 60% international exposure, 30% local equity/dividend, and 10% property. Rebalancing should occur annually.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants
START_AGE = 20
RETIRE_AGE = 65
YEARS = RETIRE_AGE - START_AGE
START_SALARY = 360_000
SALARY_GROWTH = 0.07  # 7% per year
INFLATION = 0.05  # 5% inflation
BASE_EXPENSE = 22_000 * 12  # R22k/month in year 1
HOME_COST = 1_200_000
VACATION_HOME_COST = 900_000
CAR_COST = 250_000
CHILD_EDU_COST = 60_000

# Investment parameters
INIT_PORTFOLIO = 100_000
RETURNS = 0.08  # 8% expected annual return
TAX_RATE = 0.18

# Crypto & gold allocation (added from age 32 onward)
NEW_ASSET_AGE = 32
```

```python
# Create DataFrame to hold simulation
years = list(range(START_AGE, RETIRE_AGE + 1))
df = pd.DataFrame({'Age': years})
df['Year'] = pd.date_range(start='2025', periods=len(df), freq='YE').year

# Calculate inflation-adjusted expenses
df['Inflation_Factor'] = [(1 + INFLATION) ** (i) for i in range(len(df))]
df['Salary'] = [START_SALARY * ((1 + SALARY_GROWTH) ** i) for i in range(len(df))]
df['Salary'] = df['Salary'].round(2)

# Adjusted expenses
df['Expenses'] = BASE_EXPENSE * df['Inflation_Factor']

# Savings % changes over time
df['Savings_Rate'] = np.where(df['Age'] < 35, 0.15, 0.25)
df['Savings'] = df['Salary'] * df['Savings_Rate']

# Net savings (investable amount)
df['Net_Cashflow'] = df['Salary'] - df['Expenses']
df['Investments'] = df['Savings'].copy()
```

```python
# Add fixed expenses
df['Buy_House'] = np.where(df['Age'] == 30, HOME_COST, 0)
df['Vacation_Home'] = np.where(df['Age'] == 45, VACATION_HOME_COST, 0)
df['Car_Expense'] = np.where((df['Age'] - START_AGE) % 7 == 0, CAR_COST * df['Inflation_Factor'], 0)

# Education (kids born at 30, 33, 36 ‚Äî pay from ages 48‚Äì54)
edu_ages = [48, 51, 54]
df['Child_Education'] = df['Age'].apply(lambda x: CHILD_EDU_COST if x in edu_ages else 0)

# Total lifestyle withdrawals
df['Lifestyle_Withdrawals'] = df['Buy_House'] + df['Vacation_Home'] + df['Car_Expense'] + df['Child_Education']
```

```python
portfolio = INIT_PORTFOLIO
net_worth = []

for i, row in df.iterrows():
    portfolio *= (1 + RETURNS)  # growth
    portfolio += row['Investments']  # add new investment
    portfolio -= row['Lifestyle_Withdrawals']  # subtract expenses
    net_worth.append(portfolio)

df['Portfolio_Value'] = net_worth
```

```python
final_portfolio = df['Portfolio_Value'].iloc[-1]
capital_gain = final_portfolio - INIT_PORTFOLIO
tax_due = capital_gain * TAX_RATE
after_tax_value = final_portfolio - tax_due
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Age'], df['Portfolio_Value'], label='Pre-Tax Portfolio')
plt.axhline(after_tax_value, linestyle='--', color='orange', label='After-Tax Retirement Value')
plt.title("Apollo's Projected Net Worth Over Time")
plt.xlabel("Age")
plt.ylabel("Portfolio Value (ZAR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
summary = pd.DataFrame({
    'Retirement Age': [RETIRE_AGE],
    'Final Portfolio Value (Pre-Tax)': [round(final_portfolio, 2)],
    'Capital Gains Tax (18%)': [round(tax_due, 2)],
    'Final Value (After-Tax)': [round(after_tax_value, 2)]
})
summary.T
```

## üìà Linear Regression Forecasting
We use linear regression to forecast the growth of Apollo's portfolio from the available historical data. This will help estimate the portfolio's value at age 65.

```python

from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Prepare data for regression
portfolio_value = normalized.sum(axis=1) * 100_000
portfolio_value = portfolio_value.dropna()

# Convert dates to ordinal for regression
X = np.array([d.toordinal() for d in portfolio_value.index]).reshape(-1, 1)
y = portfolio_value.values

# Train model
model = LinearRegression()
model.fit(X, y)

# Forecast to retirement (age 65 = ~45 years from now)
future_days = 45 * 365
future_date = portfolio_value.index[-1] + pd.Timedelta(days=future_days)
future_ordinal = np.array([[future_date.toordinal()]])
forecast_value = model.predict(future_ordinal)[0]

# R-squared score
r2 = model.score(X, y)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(portfolio_value.index, y, label='Historical')
plt.plot([portfolio_value.index[-1], future_date], [y[-1], forecast_value], 'r--', label='Forecast')
plt.title("Portfolio Forecast using Linear Regression")
plt.xlabel("Date")
plt.ylabel("Portfolio Value (ZAR)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Forecasted Portfolio Value at Retirement (age 65): R{forecast_value:,.2f}")
print(f"R¬≤ Score of Model: {r2:.4f}")
```

## üí∏ Capital Gains Tax Adjustment
Apply an 18% tax on the investment gain at liquidation (retirement age).

```python

# Calculate gain and tax
initial_value = 100_000
gain = forecast_value - initial_value
tax = gain * 0.18
after_tax_value = forecast_value - tax

print(f"Capital Gains: R{gain:,.2f}")
print(f"Capital Gains Tax (18%): R{tax:,.2f}")
print(f"After-Tax Portfolio Value: R{after_tax_value:,.2f}")
```

## üìâ Inflation Adjustment
Adjust the after-tax amount for expected inflation to estimate the real value at retirement.

```python

# Adjust for inflation
inflation_rate = 0.05  # 5% annual
real_value = after_tax_value / ((1 + inflation_rate) ** 45)

print(f"Real Value at Retirement (Adjusted for 5% Inflation): R{real_value:,.2f}")
```
